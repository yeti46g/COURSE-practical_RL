# OpenAI Gym

We need some decision problems to test RL algorithsm, especially MDPs.

OpenAI gym is a python library that wraps many classical decision problems including robot control, videogames, and board games. 


```python
import gym
import matplotlib.pyplot as plt
%matplotlib inline
```

## (1) Classic Control : MountainCar-v0

1. Description: get an under powered car to the top of a hill (top = 0.5 position).
2. Observation: Box(2) 0-position,[-1.2,0.6]; 1-velocity, [-0.07,0.07].
3. Actions: Discrete(3) 0-push left; 1-no push; 2-push right.
4. Reward: -1 for each time step, until the goal of position of 0.5 is reached.
5. Starting state: random position from -0.6 to -0.4 with no velocity.
6. Episode termination: the episode ends when you reach 0.5 position, or if 200 iterations are reached. 

```python
env = gym.make("MountainCar-v0")
print("Observation space: ", env.observation_space)
print("Action space: ", env.action_space)
obs0 = env.reset()
print("Initial observation code: ", obs0, ", initial position and velocity")
plt.imshow(env.render("rgb_array"))
```

```python
print("Take action 2 :(push right)")
new_obs, reward, is_done, _ = env.step(2)
print("New observation code:, ",new_obs)
print("Reward: ,", reward)
print("is game over?:", is_done)
```

```python
time_limit = 250
env = gym.make("MountainCar-v0")
s = env.reset()
actions = {'left':0, 'stop':1, 'right':2}

%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()    

def policy(s, t):
    if s[1]>=0:
        return actions['right']
    else:
        return actions['left']


for t in range(time_limit):
    
    s, r, done, _ = env.step(policy(s, t))
    ax.clear()
    ax.imshow(env.render('rgb_array'))
    fig.canvas.draw()
    
    if done:
        print("Well done!")
        break
else:    
    print("Time limit exceeded. Try again.")
```

## (3) Box2D : LunarLander-v2 

```python
env = gym.make("LunarLander-v2")
env.reset()
plt.imshow(env.render('rgb_array'))
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample()) # take a random action
env.close()
```
